```python
from matplotlib import pyplot as plt
import numpy as np

def func(x):
    return -(x / 10) + 1 - 3 * (np.sqrt(x / 8))
```

```python
#plot a beaty graph with labels
def plot_graph(x, y):
    plt.plot(x, y)
    plt.xlabel('players totais')
    plt.ylabel('players que analisam')
    plt.title('grafico de jogadores que analisam')
    plt.savefig('graph.png')
    plt.show()
    #save
```

```python
n = 1
x = np.linspace(0, n, 100)
y = func(x)
plot_graph(x, y)
```
